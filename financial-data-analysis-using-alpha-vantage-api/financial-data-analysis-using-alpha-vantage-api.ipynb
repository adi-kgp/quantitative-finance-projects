```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
```

```python
stock = "AAPL"
url = f"https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol={stock}&apikey={alpha_vantage_api_key}"
```

```python
response = requests.get(url)
```

```python
data = response.json()
```

```python
data.keys()
```

```python
d = data['Time Series (Daily)']
```

```python
df = pd.DataFrame.from_dict(d, orient='index')
df
```

```python
df.index = pd.to_datetime(df.index)
```

```python
df.index
```

```python
df.values
```

```python
df.info()
```

```python
df = df.map(pd.to_numeric)
```

```python
df.info()
```

```python
df.head()
```

```python
max_close = df['4. close'].max()
max_close
```

```python
max_close_condition = df['4. close'] == max_close
```

```python
max_close_date = df[max_close_condition]
```

```python
# plt.style.use('seaborn')
plt.plot(df.index, df['4. close'])
plt.title(f"{stock} historic prices")
plt.scatter(max_close_date.index, max_close_date['4. close'], color='red', marker="*", s=400)
plt.ylabel("USD")
plt.xlabel("Date")
plt.show()
```

```python
returns = df['4. close'].pct_change()
returns.dropna(inplace=True)
```

```python
returns
```

```python
plt.hist(returns, bins=50)
plt.title(f"{stock} Returns")
plt.ylabel("Frequency")
plt.xlabel("Returns")
plt.show()
```
