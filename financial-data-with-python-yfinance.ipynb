#### Advanced Financial Data Gathering

```python
# Major financial indices
major_indices = pd.read_html('https://finance.yahoo.com/world-indices/')[0]
```

```python
major_indices.head()
```

```python
ticker_list = major_indices['Symbol'].str.replace("^", "").str.lower().to_list()
```

```python
tickerData = yf.Tickers(ticker_list)
```

```python
tickerData
```

```python
df = yf.download(ticker_list, period='1d', start='2020-01-13', end='2021-03-10')
```

```python
len(ticker_list)
```

```python
df
```

```python
df.columns
```

```python
df.head()
```

```python
adj_close = df.dropna(thresh=10, axis=1)['Adj Close'].drop('XAX', axis=1)
```

```python
adj_close
```

```python
adj_close.describe().T
```

```python
adj_close.plot(subplots=True)
```

```python
adj_close.plot(figsize=(12,20), subplots=True)
plt.show()
```

```python
dji = adj_close['DJI']
dji
```

```python
dji.resample("4ME").mean()
```

```python
dji_per_change = dji/(dji.shift(1) -1)
```

```python
dji_per_change.plot(figsize=(12,6))
```

```python
dji_log_returns_shift = np.log(dji / dji.shift(1))
```

```python
dji_log_returns_shift.plot(figsize=(12,6));
```

```python
dji.hist(bins=50, figsize=(8,8));
```

```python
dji_log_returns_shift.hist(bins=50, figsize=(8,8));
```
